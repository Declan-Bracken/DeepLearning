```python
# EXTRACT AND SPLIT
import numpy as np
import os
import torch
import torchvision
from torchvision import transforms

#Location of dataset in colab
dataset_root = '/content/Lab_3b_Gesture_Dataset'

def make_divisible(y,x):
  """
  Function which will round down the integer y until it is divisible by integer x
  """
  for i in range(y):
  #check if y is divisible by x
    if y%x != 0:
      #if not divisible, reduce by 1 and repeat
      y -= 1
    else:
      return y

def get_data_loader(data_path, batch_size, val_split, test_split):

  """
  Args:
    data_path: path to location of the unzipped dataset in colab
    batch_size: the number of samples per batch
    val_split: Rough fraction of data to save for validation
    test_split: Rough fraction of data to save for testing

  Outputs:
    train_loader: iterable training dataset
    val_loader: iterable validation dataset
    test_loader: iterable testing dataset
    classes: list of possible classes

  """

  #create data transformation
  transform = transforms.Compose([
      transforms.Resize((224, 224)),  # Resize images to dim = 224 in case not already
      transforms.ToTensor(),  # Convert images to tensors
      transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Normalize pixel values
  ])

  #Load Data
  dataset = torchvision.datasets.ImageFolder(root=data_path, transform=transform)

  #List of Classes
  classes = dataset.classes

  samples_per_person = 3

  # Initialize lists to hold indices for train, validation, and test datasets
  train_indices = []
  val_indices = []
  test_indices = []

  #initialize total amount of data in each of the splits.
  total_train_size, total_val_size, total_test_size = 0, 0, 0

  #loop through classes to split data
  for class_idx in range(len(classes)):

    # Get liest of indices for the current class
    sample_indices = [idx for idx, label in enumerate(dataset.targets) if label == class_idx]

    #get splits, make divisible by 3 to ensure that they're split by person
    test_size = make_divisible(int(test_split * len(sample_indices)),samples_per_person)
    val_size = make_divisible(int(val_split * len(sample_indices)),samples_per_person)
    train_size = make_divisible(len(sample_indices) - test_size - val_size,samples_per_person)

    #add up net sizes for later printing
    total_train_size += train_size
    total_val_size += val_size
    total_test_size += test_size

    #split data indices
    test_indices.extend(sample_indices[:test_size]) #take first part as test
    val_indices.extend(sample_indices[test_size+1:test_size+1+val_size]) #second section as validation
    train_indices.extend(sample_indices[test_size+1+val_size:]) #everything else for training

  # Define the random seed for reproducibility
  gen = torch.Generator().manual_seed(42)

  #Create Samplers for split indices
  train_sampler = torch.utils.data.SubsetRandomSampler(train_indices, generator=gen)
  val_sampler = torch.utils.data.SubsetRandomSampler(val_indices, generator=gen)
  test_sampler = torch.utils.data.SubsetRandomSampler(test_indices, generator=gen)

  #create loader objects for output
  train_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, sampler = train_sampler)
  val_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, sampler = val_sampler)
  test_loader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, sampler = test_sampler)

  print("Train size: {} \nVal size: {} \nTest size: {}".format(total_train_size, total_val_size, total_test_size))

  return train_loader, val_loader, test_loader, classes

train_loader, val_loader, test_loader, classes = get_data_loader(dataset_root, 64, 0.1, 0.1)
```

### Part (i) Convolutional Network [2pt MODEL]

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
# TO BE COMPLETED
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
  def __init__(self):
      super(CNN, self).__init__()
      self.name = "CNN"
      #Output size = [(Input size - Kernel size + 2 * Padding) / Stride] + 1

      # 3x224x224 -> 5x74.6x74.6 -> 5x74x74 (conv1)
      self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 5, kernel_size = 5, stride = 3, padding = 1) #(5*5*3+1) * 5 = 380 params
      # 5x74x74 -> 5x37x37 (pool)
      self.pool = nn.MaxPool2d(2, 2)
      # 5x37x37 -> 10x12.3x12.3 -> 10x12x12 (conv2)
      self.conv2 = nn.Conv2d(in_channels = 5, out_channels = 10, kernel_size = 5, stride = 3, padding = 1) #(5*5*5+1) * 10 = 1510 params
      # 10x12x12 -> 10x6x6(second pool)
      # 10x6x6 = 1x360 -> 1x80 (fc1)
      self.fc1 = nn.Linear(360, 80) #28,800 params
      self.fc2 = nn.Linear(80, 9) #720 params


  def forward(self, x):
      x = self.pool(F.relu(self.conv1(x)))
      x = self.pool(F.relu(self.conv2(x)))
      x = x.view(-1, 360)
      x = self.fc1(x)
      x = self.fc2(x)
      x = x.squeeze(1)
      return x
```

### Part (ii) Training Code [2pt MODEL]

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

```python
# TO BE COMPLETED

import torch.optim as optim
from matplotlib import pyplot as plt

#same function as sample code
def get_accuracy(model, loader, use_cuda = True):
    """
    Gets model accuracy. Same as sample code.

    Args:
      model: net used for training .
      data_loader: either the training or validation set to get accuracy.
      use_cuda: Boolean set to True. Uses GPU

    Out:
      Accuracy: accuracy of the model on the loader dataset.
    """

    correct = 0
    total = 0

    for imgs, labels in loader:

        #############################################
        #To Enable GPU Usage
        if use_cuda and torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()
        #############################################

        output = model(imgs)

        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        #calculate number of correct predictions
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
    accuracy = correct / total
    return accuracy

def create_model_checkpoint(model, batch_size, learning_rate, epoch):
    """
    Creates a checkpoint of the current mode. Similar to assignment 1

    Args:
      model: net used for training
      batch_size: batch size
      learning_rate: learning rate
      epoch: Epoch of current model
    """
    #create pathname
    model_path = "model_{}_batch_{}_lr_{}_epoch_{}".format(model.name, batch_size, learning_rate, epoch)
    #save
    torch.save(model.state_dict(), model_path)
    return

#Training Function (similar to sample code, but not the same)
def train(model, train_loader, val_loader, lr=0.001, training_batch_size=64, num_epochs=1, save = True, use_cuda = True):

  """
  Very similar function to sample code.

  Args:
    model: Net to use for training and validation
    train_loader: Training dataset in the form of torch loader object
    val_loader: Validation dataset in the form of torch loader object
    lr: Learning Rate
    training_batch_size: Batch size used for training
    num_epochs: Number of epochs to loop through
    save: Boolean (default True) to toggle if you want to save model parameters
    use_cuda: Boolean set to True automatically. Uses GPU
  """

  # Use cross entropy
  criterion = nn.CrossEntropyLoss()

  # Use Adam optimizer (different from original function)
  optimizer = optim.Adam(model.parameters(), lr=lr)

  # Initialize arrays
  train_loss, train_acc, val_acc = [], [], []
  iters = []
  n = 0

  # training
  for epoch in range(num_epochs):
      n += 1
      #initialize loss and error for the current epoch
      correct = 0
      total_train_loss = 0.0
      total_train_acc = 0.0
      total_samples = 0
      iters.append(n)

      for imgs, labels in iter(train_loader):

          #To Enable GPU Usage
          if use_cuda and torch.cuda.is_available():
            imgs = imgs.cuda()
            labels = labels.cuda()

          # forward pass
          out = model(imgs)
          loss = criterion(out, labels) # compute the total loss
          loss.backward()               # backward pass (compute parameter updates)
          optimizer.step()              # make the updates for each parameter
          optimizer.zero_grad()         # a clean up step for PyTorch

          #get training loss of this batch
          total_train_loss += float(loss.item())
          #get predictions for this batch
          pred = out.max(1, keepdim=True)[1]
          #calculate number of correct predictions by comparing with labels
          correct += pred.eq(labels.view_as(pred)).sum().item()
          #get number of samples and add
          total_samples += imgs.shape[0]

      #Calculate training and validation accuracy/loss of current epoch
      train_acc.append(correct / total_samples)
      train_loss.append(total_train_loss / training_batch_size)
      val_acc.append(get_accuracy(model, val_loader))  # compute validation accuracy

      #Print updated stats
      print("epoch: {}, Training Accuracy: {}, Training Loss: {}, Validation Accuracy: {} "
      .format(epoch, train_acc[epoch], train_loss[epoch], val_acc[epoch]))

      #save if save = True
      if save:
        if epoch == (num_epochs-1):
          #labeled 'Final' for convenience
          create_model_checkpoint(model, training_batch_size, lr, 'Final')
        else:
          create_model_checkpoint(model, training_batch_size, lr, epoch)

      if train_acc[-1] == 1.0:
        print("Overfitting achieved!")
        break

  # plotting
  plt.title("Training Curve")
  plt.plot(iters, train_loss, label="Train")
  plt.xlabel("Iterations")
  plt.ylabel("Loss")
  plt.show()

  plt.title("Training Curve")
  plt.plot(iters, train_acc, label="Train")
  plt.plot(iters, val_acc, label="Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Training Accuracy")
  plt.legend(loc='best')
  plt.show()

  print("Final Training Accuracy: {}\nFinal Validation Accuracy: {}".format(train_acc[-1],val_acc[-1]))
  print("Highest Validation Accuracy: ", np.max(val_acc))
```

```python
"""
Vary number of convolutional kernels in the second layer to 15 and 8:
I would expect the network with 15 kernels to perform better as it would
extract more features after the first convolution.
"""
class CNN_20kernels(nn.Module):
  def __init__(self):
      super(CNN_20kernels, self).__init__()
      self.name = "CNN_20kernels"
      #Output size = [(Input size - Kernel size + 2 * Padding) / Stride] + 1

      # 3x224x224 -> 7x111.5x11.5 -> 111x111 (conv1)
      self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 7, kernel_size = 5, stride = 2, padding = 1) #(5*5*3+1) * 5 = 380 params
      # 7x111x111 -> 7x55x55 (pool)
      self.pool = nn.MaxPool2d(2, 2)
      # 7x55x55 -> 20x27x27 (conv2)
      self.conv2 = nn.Conv2d(in_channels = 7, out_channels = 20, kernel_size = 5, stride = 2, padding = 1) #(5*5*5+1) * 20 = 2520 params
      # 20x27x27 -> 20x13x13(second pool)
      # 20x13x13 = 1x720
      self.fc1 = nn.Linear(3380, 380)
      # self.fc1 = nn.Linear(720,380)
      self.fc2 = nn.Linear(380, 90)
      self.fc3 = nn.Linear(90, 9)


  def forward(self, x):
      x = self.pool(F.relu(self.conv1(x)))
      x = self.pool(F.relu(self.conv2(x)))
      x = x.view(-1, 3380)
      x = self.fc1(x)
      x = self.fc2(x)
      x = self.fc3(x)
      x = x.squeeze(1)
      return x

class CNN_15kernels(nn.Module):
  def __init__(self):
      super(CNN_15kernels, self).__init__()
      self.name = "CNN_15kernels"
      #Output size = [(Input size - Kernel size + 2 * Padding) / Stride] + 1

      # 3x224x224 -> 5x74.6x74.6 -> 5x74x74 (conv1)
      self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 5, kernel_size = 5, stride = 3, padding = 1) #(5*5*3+1) * 5 = 380 params
      # 5x74x74 -> 7x37x37 (pool)
      self.pool = nn.MaxPool2d(2, 2)
      # 7x37x37 -> 15x12.3x12.3 -> 15x12x12 (conv2)
      self.conv2 = nn.Conv2d(in_channels = 5, out_channels = 15, kernel_size = 5, stride = 3, padding = 1) #(5*5*5+1) * 15 = 2265 params
      # 15x12x12 -> 15x6x6(second pool)
      # 15x6x6 = 1x540
      self.fc1 = nn.Linear(540, 9) #43,200 params
      # self.fc2 = nn.Linear(120, 9) #1080 params


  def forward(self, x):
      x = self.pool(F.relu(self.conv1(x)))
      x = self.pool(F.relu(self.conv2(x)))
      x = x.view(-1, 540)
      x = self.fc1(x)
      # x = self.fc2(x)
      x = x.squeeze(1)
      return x

class CNN_15kernels2(nn.Module):
  def __init__(self):
      super(CNN_15kernels2, self).__init__()
      self.name = "CNN_15kernels2"
      #Output size = [(Input size - Kernel size + 2 * Padding) / Stride] + 1

      # 3x224x224 -> 5x74.6x74.6 -> 5x74x74 (conv1)
      self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 5, kernel_size = 5, stride = 2, padding = 1) #(5*5*3+1) * 5 = 380 params
      # 5x74x74 -> 7x37x37 (pool)
      self.pool = nn.MaxPool2d(2, 2)
      # 7x37x37 -> 15x12.3x12.3 -> 15x12x12 (conv2)
      self.conv2 = nn.Conv2d(in_channels = 5, out_channels = 15, kernel_size = 5, stride = 2, padding = 1) #(5*5*5+1) * 15 = 2265 params
      # 15x12x12 -> 15x6x6(second pool)
      # 15x6x6 = 1x540
      self.fc1 = nn.Linear(2535, 9) #43,200 params
      # self.fc2 = nn.Linear(120, 9) #1080 params


  def forward(self, x):
      x = self.pool(F.relu(self.conv1(x)))
      x = self.pool(F.relu(self.conv2(x)))
      x = x.view(-1, 2535)
      x = self.fc1(x)
      # x = self.fc2(x)
      x = x.squeeze(1)
      return x

class CNN_8kernels(nn.Module):
  def __init__(self):
      super(CNN_8kernels, self).__init__()
      self.name = "CNN_8kernels"
      #Output size = [(Input size - Kernel size + 2 * Padding) / Stride] + 1

      # 3x224x224 -> 5x74.6x74.6 -> 5x74x74 (conv1)
      self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 4, kernel_size = 5, stride = 2, padding = 1)
      # 5x74x74 -> 5x37x37 (pool)
      self.pool = nn.MaxPool2d(2, 2)
      # 5x37x37 -> 10x12.3x12.3 -> 10x12x12 (conv2)
      self.conv2 = nn.Conv2d(in_channels = 4, out_channels = 8, kernel_size = 5, stride = 2, padding = 1)
      # 8x12x12 -> 8x6x6(second pool)
      self.fc1 = nn.Linear(1352, 9)
      # self.fc2 = nn.Linear(450, 180)
      # self.fc3 = nn.Linear(180, 9)
      # self.fc1 = nn.Linear(288, 60)
      # self.fc2 = nn.Linear(60, 9)


  def forward(self, x):
      x = self.pool(F.relu(self.conv1(x)))
      x = self.pool(F.relu(self.conv2(x)))
      x = x.view(-1, 1352)
      x = self.fc1(x)
      # x = self.fc2(x)
      x = x.squeeze(1)
      # x = nn.functional.softmax(x, dim=1)
      return x
```

## Part 4. Transfer Learning [8pt]
For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal
role in establishing Deep Neural Nets as a go-to tool for image classification problems and we will use an
ImageNet pre-trained AlexNet model to extract features in this assignment.

### Part (i) - [2pt EXPLORATORY]

will download the pretrained weights from the internet.

```python
import torchvision.models
alexnet = torchvision.models.alexnet(pretrained=True)
use_cuda = True
if use_cuda and torch.cuda.is_available():
  alexnet.cuda()
  print('Alexnet on GPU ...')
```

The alexnet model is split up into two components: *alexnet.features* and *alexnet.classifier*. The
first neural network component, *alexnet.features*, is used to compute convolutional features, which are
taken as input in *alexnet.classifier*.

The neural network alexnet.features expects an image tensor of shape Nx3x224x224 as input and it will
output a tensor of shape Nx256x6x6 . (N = batch size).

Compute the AlexNet features for each of your training, validation, and test data. Here is an example code
snippet showing how you can compute the AlexNet features for some images (your actual code might be
different):

```python
# img = ... a PyTorch tensor with shape [N,3,224,224] containing hand images ...
def extract_features(model, data_loader):

    features = []
    labels = []

    with torch.no_grad():
        for inputs, targets in data_loader:
            if use_cuda and torch.cuda.is_available():
              inputs = inputs.cuda()
              targets = targets.cuda()
            features_batch = model.features(inputs)
            features.append(features_batch)
            labels.append(targets)
    feats, labels = torch.cat(features), torch.cat(labels)
    print("Tensor Size:",feats.size())
    return feats, labels

batch_size = 1942 #entire training set
val_split = 0.1
test_split = 0.1

train_loader, val_loader, test_loader, _ = get_data_loader(dataset_root, batch_size, val_split, test_split)

train_features, train_labels = extract_features(alexnet, train_loader)
val_features, val_labels = extract_features(alexnet, val_loader)
test_features, test_labels = extract_features(alexnet, test_loader)
```
